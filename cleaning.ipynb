```python
import pandas as pd
import numpy as np
rawdata = pd.read_csv("Speed_Dating_Data.csv", encoding = "ISO-8859-1")
rawdata
```

Removing unimportant columns

```python
trunci = rawdata.columns.get_loc("attr")
truncdata = rawdata.iloc[:,range(trunci)]
truncdata = truncdata.drop(["id","idg","condtn","wave",
                            "round","position","positin1","order","partner",
                            "match","int_corr","samerace","age_o","race_o",
                            "pf_o_att","pf_o_sin","pf_o_int","pf_o_fun","pf_o_amb","pf_o_sha",
                            "dec_o","attr_o","sinc_o","intel_o","fun_o","amb_o","amb_o","shar_o",
                            "like_o","shar_o","like_o","prob_o","met_o",
                            "undergra", "mn_sat", "tuition",
                            "from", "zipcode","income",
                            "field", "career", "career_c",
                            "exphappy", "expnum"]
                           ,axis=1)
truncdata
```

```python
for v in ["attr","sinc","intel","fun","amb"]:
    truncdata[v+"3_1"] = np.nansum(truncdata.loc[:,[v+"3_1",v+"5_1"]],axis=1)
    truncdata.loc[:,[v+"3_1"]]
    truncdata = truncdata.drop([v+"5_1"],axis=1)
```

```python
for c in ["attr4_1","sinc4_1","intel4_1","fun4_1","amb4_1","shar4_1"]:
    for g in [0,1]:
        truncdata.loc[truncdata["gender"] == g, c] = truncdata.loc[truncdata["gender"] == g, c].fillna(truncdata.loc[truncdata["gender"] == g, c].mean())
```

```python
truncdata.dropna(inplace=True)
```

```python
truncdata.loc[truncdata["iid"]==12,:]
```

```python
D = dict()
for index, row in truncdata.iterrows():
    iid = row["iid"]
    if iid not in D:
        D[iid] = [row.drop(["iid","pid","dec"]),[]]
    D[iid][1].append((row["pid"],row["dec"]))
```

```python
newdf = pd.DataFrame()
for iid, v in D.items():
    for pid, dec in v[1]:
        try:
            newrow = list(D[iid][0])+list(D[pid][0])+[dec]
        except:
            continue
        newdf = newdf.append([newrow], ignore_index=True)
```
