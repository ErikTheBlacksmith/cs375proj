```python
for v in ["attr","sinc","intel","fun","amb"]:
    truncdata[v+"3_1"] = np.nansum(truncdata.loc[:,[v+"3_1",v+"5_1"]],axis=1)
    truncdata.loc[:,[v+"3_1"]]
    truncdata = truncdata.drop([v+"5_1"],axis=1)
```

```python
for c in ["attr4_1","sinc4_1","intel4_1","fun4_1","amb4_1","shar4_1"]:
    for g in [0,1]:
        truncdata.loc[truncdata["gender"] == g, c] = truncdata.loc[truncdata["gender"] == g, c].fillna(truncdata.loc[truncdata["gender"] == g, c].mean())
```

```python
D = dict()
for index, row in truncdata.iterrows():
    iid = row["iid"]
    if iid not in D:
        D[iid] = [row.drop(["iid","pid","dec"]),[]]
    D[iid][1].append((row["pid"],row["dec"]))
```

```python
newdf = pd.DataFrame()
for iid, v in D.items():
    for pid, dec in v[1]:
        try:
            newrow = list(D[iid][0])+list(D[pid][0])+[dec]
        except:
            continue
        newdf = newdf.append([newrow], ignore_index=True)
```
