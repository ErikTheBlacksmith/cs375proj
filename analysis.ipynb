```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("Cleaned_Data.csv", encoding = "ISO-8859-1")

categorical_heads = ["gender","field_cd","race"]
for col in categorical_heads + [ch+"o" for ch in categorical_heads]:
    data[col] = data[col].astype("category")

data.drop([col for col in data.columns if "2" in col or "4" in col], axis=1,inplace=True)

mdata = data.loc[np.equal(data.iloc[:,0],1),:]
fdata = data.loc[np.equal(data.iloc[:,0],0),:]
```

```python
# total correlation
plt.subplots(figsize=(20,15))
ax = plt.axes()
ax.set_title("Correlation Heatmap")
corr = data.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

Correlation between features and their decision

```python
s = corr["dec"].sort_values()[:-1]
s_s = s[[ind[-1]!='o' for ind in s.index]]
s_o = s[[ind[-1]=='o' for ind in s.index]]
#s.plot(kind="barh",figsize=(20,15), color="g")
```

```python
s_s.plot(kind="barh",figsize=(10,7), color="g")
```

```python
s_o.plot(kind="barh",figsize=(10,7), color="g")
```
